```python
import pandas as pd
```

```python
x=[[5],[8],[10],[12],[15],[3],[7],[9],[14],[6]]
y=[55,65,70,75,85,50,60,72,80,58]
```

```python
x=pd.DataFrame(x)
y=pd.Series(y)

print(x.shape)
print(y.shape)
```

```python
print(x.corrwith(y,method='pearson'))
print(x.corrwith(y,method='spearman'))
```
