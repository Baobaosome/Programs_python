```python
import pandas as pd
```

```python
s1=pd.Series({"小明":92,"小红":67,"小杰":70,"小丽":88,"小华":76})
s2=pd.Series({"小明":95,"小杰":85,"小宇":60,"小娟":79,"小彤":76})
print(s1)
print(s2)
```

```python
s1.add(s2,fill_value=0)
```

```python
print(s1.max())
print(s1.min())
print(s1.sum())
print(s1.mean())
s1.describe()
```

```python
s1=s1+5
s1
```

```python
def get_grade(score):
    if score>=90:
        return "A"
    elif score>=80:
        return "B"
    elif score>=70:
        return "C"
    elif score>=60:
        return "D"
    else: return "不及格"    
```

```python
grade=s1.apply(get_grade)
grade
```
